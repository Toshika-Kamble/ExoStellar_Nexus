```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/image.jpg')
cv2_imshow(img)
cv2.waitKey(0)
```

```python
kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
img_sharp = cv2.filter2D(img, -1, kernel)
# img_sharp = cv2.cvtColor(img_sharp, cv2.COLOR_RGB2BGR)
cv2_imshow(img_sharp)
cv2.waitKey(0)

kernel = np.array([[0,-1,0], [-1,5,-1], [0,-1,0]])
img_sharp = cv2.filter2D(img_sharp, -1, kernel)
cv2_imshow(img_sharp)
cv2.waitKey(0)
```

```python
gray = cv2.cvtColor(img_sharp, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
cv2.waitKey(0)
```

```python
ret, thresh1 = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
cv2_imshow(thresh1)
cv2.waitKey(0)
```
